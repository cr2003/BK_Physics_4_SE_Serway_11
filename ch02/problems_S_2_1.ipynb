```python
def fmt_vec(v, precision=2, type="f"):
    """Format the output of a vector with the specified precision"""
    return f"<{v.x:.{precision}{type}}, {v.y:.{precision}{type}}, {v.z:.{precision}{type}}>"
```

#### Problem 1

page 40

Two points in the xy plane have Cartesian coordinates
(2.00, -4.00) m and (-3.00, 3.00) m. Determine

**(a)**

The distance between these points 



```python
A = vec(2, -4, 0)  # m
B = vec(-3, 3, 0)  # m
d = A - B
print(f"The distance between A and B is {mag(d):.2f} m")
```

**(b)**

Their polar coordinates.

```python
r_A = mag(A)
theta_A_rad = atan2(A.y, A.x)
theta_A_deg = degrees(theta_A_rad)

r_B = mag(B)
theta_B_rad = atan2(B.y, B.x)
theta_B_deg = degrees(theta_B_rad)

print(f"A {A} m")
print(f"A ({r_A:.2f} m, {theta_A_deg:.1f} deg>")
print(f"B {B} m")
print(f"B ({r_B:.2f} m, {theta_B_deg:.1f} deg)")
```

#### Problem 2

Two points in a plane have polar coordinates (2.50 m, 30.0°) and 
(3.80 m, 120.0°). Determine 

**(a)** 

The Cartesian coordinates of these points 

```python
r_1 = 2.50  # m
theta_1_deg = 30.0  # deg
theta_1_rad = radians(theta_1_deg)

r_2 = 3.80  # m
theta_2_deg = 120.0  # deg
theta_2_rad = radians(theta_2_deg)


x_1 = r_1 * cos(theta_1_rad)
y_1 = r_1 * sin(theta_1_rad)
vec_1 = vec(x_1, y_1, 0)

x_2 = r_2 * cos(theta_2_rad)
y_2 = r_2 * sin(theta_2_rad)
vec_2 = vec(x_2, y_2, 0)

print(f"Vector 1 {fmt_vec(vec_1)} m")
print(f"Vector 2 {fmt_vec(vec_2)} m")
```

**(b)** 
The distance between them.

```python
d = mag(vec_1 - vec_2)
print(f"Distance d = {d:.2f} m")
```

#### Problem 3

The polar coordinates of a certain point are (r = 4.30 cm,

Find its Cartesian coordinates x and y. 



```python
r = 4.30  # cm
theta_deg = 214  # deg
theta_rad = radians(theta_deg)

x_1 = r * cos(theta_rad)
y_1 = r * sin(theta_rad)
vec_1 = vec(x_1, y_1, 0)

print(f"Vector 1 {fmt_vec(vec_1)} cm")
```

Find the polar coordinates of the points with Cartesian coordinates

```python
x = -x_1
y = y_1
theta = degrees(atan2(y, x))
r = sqrt(x**2 + y**2)
print(f"Polar coordinates ({r:.2f} cm, {theta + 360:.0f} deg>")
```
