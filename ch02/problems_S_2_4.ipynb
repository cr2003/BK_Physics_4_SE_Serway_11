```python
def fmt_vec(v, precision=2, type="f"):
    """Format the output of a vector with the specified precision"""
    return f"<{v.x:.{precision}{type}}, {v.y:.{precision}{type}}, {v.z:.{precision}{type}}>"
```

A minivan travels straight north in the right lane of a
divided highway at 28.0 m/s. A camper passes the minivan
and then changes from the left lane into the right
lane. As it does so, the camper’s path on the road is a
straight displacement at 8.50° east of north. To avoid
cutting off the minivan, the north–south distance
between the camper’s back bumper and the minivan’s

Can the camper be driven to satisfy this requirement?

A person walks 25.0° north of east for 3.10 km. How far
would she have to walk due north and due east to arrive
at the same location?

```python
r = 3.10 * u.km
theta = radians(25.0)

north = r * sin(theta)
east = r * cos(theta)
print(f"Due north = {north:.2}, due east = {east:.2f}")
```

#### Problem 13

page 41

Your dog is running around the grass in your back yard.
It undergoes successive displacements 3.50 m south,
8.20 m northeast, and 15.0 m west. What is the resultant
displacement?

```python
# First displacement
A = vec(0, -3.50, 0)
# Second displacement
r = 8.20
theta = radians(45)
B = vec(r * cos(theta), r * sin(theta), 0)
# Third displacement
C = vec(-15.0, 0, 0)

# Total displacement
d = A + B + C
r = mag(d)
theta_d = degrees(atan2(d.y, d.x))
print(
    f"The magnitud of displacement is r = {r:.2f} m, direction theta = {theta_d:.0f} deg"
)
```

The helicopter view in Fig. P2.15 shows two people pulling on a stubborn mule. 
The person on the right pulls with a force $\vec{F}_1$ of magnitude 120 N and 
direction of $\theta = 60.0°$. The person on the left pulls with a force 
$\vec{F}_2$ of magnitude 80.0 N and direction of  $\theta = 75.0°$. 
The forces are measured in units of newtons (symbolized N).

Find

**(a)** 

the single force that is equivalent to the two forces shown 





```python
F1_mag = 120
theta_1 = radians(60)
F2_mag = 80
theta_2 = radians(180 - 75)

F1 = vec(F1_mag * cos(theta_1), F1_mag * sin(theta_1), 0)
F2 = vec(F2_mag * cos(theta_2), F2_mag * sin(theta_2), 0)
F = F1 + F2
theta_F = degrees(atan2(F.y, F.x))
print(f"The result vector force F = {fmt_vec(F, 1)} N")
print("In terms of magnitude and direction:")
print(f"F({F.mag:.0f} N, {theta_F:.1f} deg)")
```

**(b)**

the force that a third person would have to exert on the mule to make the
resultant force equal to zero. 

```python
F3 = -F
print(f"The vector force F3 = {fmt_vec(F3, 1)} N")
print("In terms of magnitude and direction:")
print(f"F3({F3.mag:.0f} N, {theta_F + 180:.1f} deg)")
```

A snow-covered ski slope makes an angle of 35.0° with the horizontal.

up to a maximum displacement of 1.50 m at 16.0° from the vertical in
the uphill direction as shown in Figure P2.16. Find the components of its 
maximum displacement 

**(a)** 

parallel to the surface.

**Solution**

<img src="..//pics/FigPS2_16.png" width=200> 




```python
# Getting the x-axis along the slope downhill
# the y-axis is perpendicular to the slope, 35 deg to the vertical
r = 1.50 * u.m
theta = radians(90 + 35 + 16)

# The component parallel to the surface, x
x = r * cos(theta)
print(f"The component parallel to the surface, x = {x:.2f}")
print("The sign - indicates that the direction is toward the top of the hill.")
```

```python
# The component perpendicular to the surface, y
y = r * sin(theta)
print(f"The component perpendicular to the surface, y = {y:.3f}")
print("The direction is away from the snow.")
```
