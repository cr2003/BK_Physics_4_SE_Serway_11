You are spending the summer as an assistant learning
how to navigate on a large ship carrying freight
across Lake Erie. One day, you and your ship are to
travel across the lake a distance of 200 km traveling
due north from your origin port to your destination
port. Just as you leave your port of origin, the navigation
electronics go down. The captain continues sailing,
claiming they can depend on their years of experience
on the water as a guide. The engineers work on
the navigation system while the ship continues to sail,
and winds and waves push it off course. Eventually,
enough of the navigation system comes back up to
tell you your location. The system tells you that your
current position is 50.0 km north of the port of origin
and 25.0 km east of the port. The captain is a little
embarrassed that their ship is so far off course and
barks an order to you to tell them immediately what
heading they should set from your current position to
the destination port. Give them an appropriate heading
angle.


```python
# Destination vector
d = vec(0, 200, 0)
# Position vector
p = vec(25.0, 50.0, 0)
# course vector
c = d - p
# angle from x-axis
theta = degrees(atan2(c.y, c.x))
# In navigation angles - bearing (beta) are mesured from North clockwise
# To convert beta = 90 - theta (if the angle is negative +360)
beta = 90 - theta
if beta < 0:
    beta += 360
print(f"The naviational angle (bearing): {beta:.1f} deg")
```

Ecotourists use their global positioning system indicator
to determine their location inside a botanical
garden as latitude 0.002 43 degrees south of the equator,
longitude 75.642 38 degrees west. They wish to
visit a tree at latitude 0.001 62 degrees north, longitude
75.644 26 degrees west. (a) Determine the straightline
distance and the direction in which they can walk
to reach the tree as follows. First model the Earth
as a sphere of radius 6.37 3 106 m to determine the
westward and northward displacement components
required, in meters. Then model the Earth as a flat
surface to complete the calculation. (b) Explain why
it is possible to use these two geometrical models
together to solve the problem.





```python
import astropy.units as u
import numpy as np

# Given constants
R_earth = 6.37e6 * u.m

# Coordinates
lat_start = -0.00243 * u.deg  # South is negative
lat_end = 0.00162 * u.deg  # North is positive

lon_start = 75.64238 * u.deg  # West
lon_end = 75.64426 * u.deg  # Further West

# (a) Part 1: Compute Displacement Components
delta_lat = lat_end - lat_start
delta_lon = lon_end - lon_start

# Linear distance components via arc-length formula: s = R * theta (in radians)
dy_north = R_earth * delta_lat.to(u.rad).value
dx_west = R_earth * delta_lon.to(u.rad).value

# Part 2: Flat Surface Calculation (Pythagorean theorem)
distance = np.sqrt(dx_west**2 + dy_north**2)
angle_north_of_west = np.degrees(np.arctan2(dy_north.value, dx_west.value)) * u.deg

print(f"--- (a) Displacement Components ---")
print(f"Northward displacement (dy): {dy_north:.2f}")
print(f"Westward displacement (dx):  {dx_west:.2f}")
print(f"\n--- Straight-line Magnitude & Direction ---")
print(f"Straight-line distance:      {distance:.2f}")
print(f"Direction:                   {angle_north_of_west:.2f}° North of West")
```

You are working as an assistant to an air-traffic controller
at the local airport, from which small airplanes
take off and land. Your job is to make sure that airplanes
are not closer to each other than a minimum
safe separation distance of 2.00 km. You observe two
small aircraft on your radar screen, out over the ocean
surface. The first is at altitude 800 m above the surface,
horizontal distance 19.2 km, and 25.0° south of west.
The second aircraft is at altitude 1 100 m, horizontal
distance 17.6 km, and 20.0° south of west. Your supervisor
is concerned that the two aircraft are too close
together and asks for a separation distance for the two
airplanes. (Place the x axis west, the y axis south, and
the z axis vertical.)

```python
# Forget about book sugestion of change axis, keep standar vpython axis:
# x(-west, +east), y(+north, -south), z(+heigth)
theta1 = radians(180 + 25.0)
theta2 = radians(180 + 20.0)
r1 = 19.2
r2 = 17.6
h1 = 0.8  # km
h2 = 1.1  # km

pos1 = vec(r1 * cos(theta1), r1 * sin(theta1), h1)
pos2 = vec(r2 * cos(theta2), r2 * sin(theta2), h2)

separation_vector = pos2 - pos1
separation_distance = separation_vector.mag

print("--- Air Traffic Control Radar Report ---")
print(f"Aircraft 1 Position (West, South, Alt) km: {pos1}")
print(f"Aircraft 2 Position (West, South, Alt) km: {pos2}")
print(f"\nCalculated Separation Distance: {separation_distance:.3f} km")

if separation_distance < 2.00:
    print("\n⚠️ WARNING: Aircraft are closer than the 2.00 km safety limit!")
else:
    print("\n✅ SAFE: Separation distance meets safety requirements.")
```
