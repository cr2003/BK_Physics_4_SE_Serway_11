Given $\vec{M} = 2i - 3j + k$ and $\vec{N} = 4i + 5j - 2k$ calculate
the vector product  $\vec{M} \times \vec{N}$

The displacement vectors 42.0 cm at 15.0° and 23.0 cm at
65.0° both start from the origin and form two sides of a 
parallelogram. Both angles are measured counterclockwise
from the x axis. 

**(a)** 

Find the area of the parallelogram.

**NB:** Counterclockwise is the american term for anticlockwise.

```python
r1 = 42.0 * u.cm
theta_1 = radians(15.0)
r2 = 23.0 * u.cm
theta_2 = radians(65.0)

# The area of the parallelogram is equalt to the magnitude of the vector product
# (cross product).

area = abs(r1 * r2 * sin(theta_1 - theta_2))
print(f"Area of the parallelogram: {area:.0f}")
```

(b) Find the length of its longer diagonal.

```python
# The longer diagonal is formed by the magnitude of the sum vector (r1 + r2)
r1 = vec(42.0 * cos(radians(15.0)), 42.0 * sin(radians(15.0)), 0)
r2 = vec(23.0 * cos(radians(65.0)), 23.0 * sin(radians(65.0)), 0)

# Diagonal
diag_long = mag(r1 + r2)
print(f"Longer diagonal: {diag_long:.1f}")
```
