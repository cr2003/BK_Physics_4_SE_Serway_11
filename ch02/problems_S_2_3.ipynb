```python
def fmt_vec(v, precision=2, type="f"):
    """Format the output of a vector with the specified precision"""
    return f"<{v.x:.{precision}{type}}, {v.y:.{precision}{type}}, {v.z:.{precision}{type}}>"
```

Vector $\vec{A}$ has a magnitude of 29 units and points in the

resultant vector $\vec{A} + \vec{B}$ points in the negative y direction

A force $\vec{F}_1$ of magnitude 6.00 units
acts on an object at the origin in
a direction $\theta =  30.0°$ above the
positive x axis (Fig. P2.7). A second
force $\vec{F}_2$ of magnitude 5.00 units
acts on the object in the direction
of the positive y axis. Find graphically
the magnitude and direction
of the resultant force $\vec{F}_1$ + $\vec{F}_2$.

```python
F_1_mag = 6
theta_deg = 30
theta_rad = radians(30)
F_1 = vec(F_1_mag * cos(theta_rad), F_1_mag * sin(theta_rad), 0)

F_2_mag = 5
F_2 = vec(0, F_2_mag, 0)

F = F_1 + F_2
print(f"Resultant vector F =  {fmt_vec(F)}")
theta_F_deg = degrees(atan2(F.y, F.x))
print(f"F ({mag(F):.1f} N, {theta_F_deg:.0f} deg)")
```

The displacement vectors $\vec{A}$ and $\vec{B}$ shown in Figure P2.9
both have magnitudes of 3.00 m. The direction of vector $\vec{A}$ is
$\theta =  30.0°$. 

(Report all angles counterclockwise from the positive x axis.)

Find graphically 

**(a)**

$\vec{A} + \vec{B}$


```python
# Vector A: Magnitude 3.00 m, Angle 30 degrees
mag_A = 3
theta_A = radians(30.0)
A = vec(mag_A * cos(theta_A), mag_A * sin(theta_A), 0)

mag_B = 3
B = vec(0, mag_B, 0)

R = A + B
print(f"Resultant vector R =  {fmt_vec(R)}")
theta_R_deg = degrees(atan2(R.y, R.x))
print(f"R({mag(R):.1f} m, {theta_R_deg:.0f} deg)")
```

**(b)**

$\vec{A} - \vec{B}$

```python
R = A - B
print(f"Resultant vector R =  {fmt_vec(R)}")
theta_R_deg = degrees(atan2(R.y, R.x))

# They ask to report theta as counterclockwise from positive x axis
theta = 360 + theta_R_deg
print(f"R({mag(R):.1f} m, {theta:.0f} deg)")
```

**(c)**

$\vec{B} - \vec{A}$

```python
R = B - A
print(f"Resultant vector R =  {fmt_vec(R)}")
theta_R_deg = degrees(atan2(R.y, R.x))

print(f"R({mag(R):.1f} m, {theta_R_deg:.0f} deg)")
```

**(d)**

$\vec{A} - 2\vec{B}$

```python
R = A - 2 * B
print(f"Resultant vector R =  {fmt_vec(R)}")
theta_R_deg = degrees(atan2(R.y, R.x))

# They ask to report theta as counterclockwise from positive x axis
theta = 360 + theta_R_deg
print(f"R({mag(R):.1f} m, {theta:.0f} deg)")
```

#### Problem 10

A roller-coaster car moves 200 ft horizontally and then rises 135 ft at an 
angle of 30.0° above the horizontal. It next travels 135 ft at an angle of 
40.0° downward. What is its displacement from its starting point? Use graphical
techniques.

**Solution**

<img src="..//pics/FigP2_10.png" width=600> 


```python
mag_A = 200  # ft
A = vec(mag_A, 0, 0)
A

# Vector B: Magnitude 135 ft, Angle 30 degrees
mag_B = 135  # ft
theta_B = radians(30.0)
B = vec(mag_B * cos(theta_B), mag_B * sin(theta_B), 0)

# Vector C: Magnitude 135 ft, Angle -40 degrees
mag_C = 135  # ft
theta_C = radians(-40.0)
C = vec(mag_C * cos(theta_C), mag_C * sin(theta_C), 0)

# Displacement
D = A + B + C
print(f"Resultant displacement D =  {fmt_vec(D)}")
theta_D_deg = degrees(atan2(D.y, D.x))

print(f"D({mag(D):.2f} ft, {theta_D_deg:.2f} deg)")
```
