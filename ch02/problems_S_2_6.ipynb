#### Problem 34

page 43

Vector $\vec{A}$ has a magnitude of 5.00 units, and vector $\vec{B}$
has a magnitude of 9.00 units. The two vectors make an
angle of 50.0° with each other. Find $\vec{A} \cdot \vec{B}$ .




```python
A_mag = 5
B_mag = 9
theta = radians(50)

# A . B = |A| |B| cos(theta)
dot_product_AB = A_mag * B_mag * cos(theta)
print(f"The Dot product AB = {dot_product_AB:.1f}")
```

#### Problem 36

page 43

Using the definition of the scalar product, find the angles between 

**(a)**

$\vec{A} = 3i - 2j$ and $\vec{B} = 4i - 4j$



```python
A = vec(3, -2, 0)
B = vec(4, -4, 0)

# Definition: dot_prod = A_mag * B*mag * cos(theta)
# ==> theta = acos(dot_prod / (A_mag * B_mag))

dot_prod = dot(A, B)
theta = degrees(acos(dot_prod / (A.mag * B.mag)))

# Check with built-in function
theta_check = degrees(diff_angle(A, B))

print(f"The angle is {theta:.1f} deg.")
print(f"With the built-in function (diff_angle), the angle is {theta_check:.1f} deg.")
```
