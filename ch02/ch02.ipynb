## 2.6 The Scalar Product of Two Vectors
page 36

<img src="..//pics/Fig2_20.png" width=400> 


The scalar product of two vector, aka dot product, is an scalar.

$$\vec{A} \cdot \vec{B} \approxeq A B \cos \theta$$

**Commutative propery**

$$\vec{A} \cdot \vec{B} = \vec{B} \cdot \vec{A}$$

**Distributive propery**

$$\vec{A} \cdot (\vec{B} + \vec{C})= \vec{A} \cdot \vec{B} + \vec{A} \cdot \vec{C}$$

The scalar product could also be expressed in its components parts:

$$\vec{A} \cdot \vec{B} = A_x B_x + A_y B_y + A_z B_z$$







### Example 2.6
page 14

The Scalar Product
The vectors $\vec{A}$ and $\vec{B}$ are given by $\vec{A} = 2 \hat{i} + 3 \hat{j}$ 
and  $\vec{B} = - \hat{i} + 2 \hat{j}$.

**(A)**

Determine the scalar product $\vec{A} \cdot \vec{B}$ .


```python
# Define two vectors
A = vec(2, 3, 0)
B = vec(-1, 2, 0)

# calculate the dot product
result_dot = dot(A, B)
print(f"Dot Product: {result_dot}")
```

**(B)**

Find the angle $\theta$ between $\vec{A}$ and $\vec{B}$

```python
# dot(A, B) = mag(A) * mag(B) * cos(theta) ==> theta = arccos (dot(A, B) / mag(a) * mag(b))
theta = acos(dot(A, B) / (mag(A) * mag(B)))
theta_deg = degrees(theta)
print(f"The angle theta = {theta_deg:.1f} deg.")
```

```python
# The professional way using diff_angle()
angle = diff_angle(A, B)  # returns value in radians
angle_deg = degrees(angle)
print(f"The angle theta = {angle_deg:.1f} deg.")
```

## 2.7 The Vector Product of Two Vectors
page 38

<img src="..//pics/Fig2_21.png" width=400> 

The vector product of two vector, aka cross product, is another vector.

$$\vec{C} = \vec{A} \times \vec{B}$$

its magnitude is

$$C = A B \sin \theta$$

The quantity $A B \sin \theta$ is equal to the parallelogram formed by $\vec{A}$
and $\vec{B}$. The direction of $\vec{C}$ is perpendicular to the plane formed by 
$\vec{A}$ and $\vec{B}$, and the best way to determine this direction is to use
the right hand rule.




**Some properties**

1. Unlike the scalar product, the vector product is not commutative. Instead the order in which the two vectors are
multiplied in a vector product is important:

$$ \vec{A} \times \vec{B} = -\vec{B} \times \vec{A}$$

2. If $\vec{A}$ is parallel to $\vec{B} (\theta = 0 \; or \; 180^\circ)$, then $ \vec{A} \times \vec{B} = 0$;
therefore, it follows that $\vec{A} \times \vec{A} = 0$

3. If $\vec{A}$ is perpendicular to $\vec{B}, then |\vec{A} \times \vec{B}| = A B$.

4. The vector product obeys the distributive law:

$$\vec{A} \times (\vec{B} + \vec{C})= \vec{A} \times \vec{B} + \vec{A} \times \vec{C}$$

5. The derivative of the vector product with respect to some variable such as t is

$$\frac{d}{dt}(\vec{A} \times \vec{B}) = \frac{d \vec{A}}{dt} \times \vec{B} + \vec{A} 
\times \frac{d \vec{B}}{dt} $$

### Think-Pair-Share

#### 1.

You are working at a radar station for the Coast Guard.
While everyone else is out to lunch, you receive a distress
call from a sinking ship. The ship is located at a
distance of 51.2 km from the station, at a bearing of
36° west of north. The locations of four other ships are
shown on your radar screen as follows:

<img src="..//pics/Table_2_1.png" width=600> 

Quick! Which ship do you contact to help the sinking
ship? Which ship will get there in the shortest
time interval? Assume that each ship would accelerate
quickly to its maximum speed and then maintain that
constant speed in a straight line for the entire trip to the
sinking ship.

```python
from vpython import vec

# 1. Define the sinking ship's position as a vector
# Angle is 36 degrees West of North, so to use standard math angle from x, angle (90 + 36)
dist_sinking = 51.2
bearing_sinking = radians(36)
theta_sinking = radians(90 + 36)

theta_sinking = radians(90 + 36)
# x = r*cos(theta), y = r*sin(theta)
pos_sinking = vec(
    dist_sinking * cos(theta_sinking), dist_sinking * sin(theta_sinking), 0
)

# 2. Data for the other ships: [Distance, 90 + Bearing_WoN, Max_Speed]
ships_data = {
    1: [36.1, 90 + 42, 30.0],
    2: [37.3, 90 + 61, 38.0],
    3: [10.2, 90 + 36, 32.0],
    4: [51.2, 90 + 79, 45.0],
}

results = []

print(f"{'Ship #':<10} | {'Distance (km)':<15} | {'Time (hours)':<15}")
print("-" * 45)

for ship_id, data in ships_data.items():
    dist, theta_deg, speed = data
    theta_rad = radians(theta_deg)

    # Define current ship position vector
    pos_ship = vec(dist * cos(theta_rad), dist * sin(theta_rad), 0)

    # 3. Vector Analysis: The displacement vector from ship to sinking victim
    # displacement = destination - start
    displacement = pos_sinking - pos_ship

    # The magnitude (length) of the displacement vector
    distance_to_travel = mag(displacement)

    # Calculate time (t = d / v)
    travel_time = distance_to_travel / speed
    results.append((travel_time, ship_id))

    print(f"{ship_id:<10} | {distance_to_travel:<15.2f} | {travel_time:<15.3f}")

# 4. Determine the winner
results.sort()  # Sorts by the first element (time)
best_time, best_ship = results[0]

print("-" * 45)
print(
    f"CONCLUSION: Contact Ship #{best_ship}. Arrival in {best_time * 60:.1f} minutes."
)
```
