## Chapter 3 - Motion in One Direction

### Summary

#### average velocity

$$v_{x,avg} \equiv \frac{\Delta x}{\Delta t}$$

#### average speed

$$v_{avg} \equiv \frac{d}{\Delta t}$$

#### instantaneous velocity

$$v_{x} \equiv \lim\limits_{\Delta t \to 0}\frac{\Delta x}{\Delta t} = \frac{dx}{dt}$$

#### average acceleration

$$a_{x,avg} \equiv \frac{\Delta v_x}{\Delta t} = \frac{v_{xf} - v_{xi}}{t_f - t_i}$$

#### instantaneous acceleration

$$a_{x} \equiv \lim\limits_{\Delta t \to 0}\frac{\Delta v_x}{\Delta t} = \frac{dv_x}{dt}$$


#### Particle under constant velocity

$$v_{x} = \frac{\Delta x}{\Delta t}$$

$$x_f = x_i + v_x t$$

#### Particle under constant speed

$$v= \frac{d}{\Delta t}$$

#### Particle under constant acceleration

$$v_{xf} = v_{xi} + a_x t$$

$$v_{x,avg} = \frac{v_{xi} + v_{xf}}{2}$$

$$x_{f} = x_i + \frac{1}{2} \left( v_{xi} + v_{xf} \right) t$$

$$x_{f} = x_i + v_{xi} t +  \frac{1}{2} a_x t^2$$

$$v_{xf}^2 = v_{xi}^2  + 2a_x \left( x_f - x_i \right)$$

## 4.4 Analysis Model: Particle in Uniform Circular Motion

page 91

**Centripetal acceleration**

An acceleration of this nature is called a `centripetal acceleration` (centripetal means
center-seeking). 

$$a_c = \frac{v^2}{r}$$

**Period**

Is the time interval required for one complete revolution. That's it the number of seconds for one revolution.

$$T = \frac{2 \pi r}{v}$$

**Angular speed**
Because one full revolution of the particle around the circle corresponds to an angle of $2\pi$ radians, the product of $2\pi$ and the rotation rate gives the `angular speed` $\omega$ of the particle, measured in radians/s or $s^{-1}$

$$\omega = \frac{2\pi}{T}$$

Combining this equation with the equation of the period, we find a relationship between angular speed and the translational speed.

$$v = r\omega$$

$$a_c = r\omega^2$$

### Example of Solving a problem with SymPy and AstroPy

#### Problem 17

page 74

An object moving with uniform acceleration has a
velocity of 12.0 cm/s in the positive x direction when its
x coordinate is 3.00 cm. If its x coordinate 2.00 s later is
-5.00 cm, what is its acceleration?

```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
(x_i, x_f, v_i, t, a) = sp.symbols("x_i, x_f, v_i, t, a", positive=True)

# Define the kinematic equation: x_f = x_i + v_i * t + 1/2 * a * t^2
equation = sp.Eq(x_f, x_i + v_i * t + sp.Rational(1, 2) * a * t**2)

# Solve the equation for the unknown acceleration 'a'

solutions = sp.solve(equation, a)
a_isolated = solutions[0]

print("--- acceleration Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(a, a_isolated))}"))
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
v_i_val = 12.0 * (u.cm / u.s)
x_i_val = 3.00 * u.cm
t_val = 2.00 * u.s
x_f_val = -5.00 * u.cm

# Convert the SymPy expression to a callable function
a_func = sp.lambdify((x_i, x_f, v_i, t), a_isolated)

# Calculate numerical value
a_val = a_func(x_i_val, x_f_val, v_i_val, t_val)

print("--- AstroPy Numerical Result ---")
print(f"Acceleration a: {a_val}")
```
