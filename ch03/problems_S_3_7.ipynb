#### Problem 14

An electron in a cathode-ray tube accelerates uniformly
from $2.00 \times 10^4$ m/s to $6.00 \times 10^6$ m/s over 1.50 cm.

**(a)** 

In what time interval does the electron travel this
1.50 cm? 

```python
vi = 2.00e4 * (u.m / u.s)
vf = 6.00e6 * (u.m / u.s)
x = 1.50 * u.cm
x = x.to(u.m)

v_avg = (vi + vf) / 2
# v = x / t ==> t = v / x
t = x / v_avg
print(f"The time interval is {t:.2e} (or {t.to(u.ns):.2f})")
```

**(b)** 

What is its acceleration?



```python
# vf**2 = vi**2 + 2 * a * x ==> a = (vf**2 - vi**2) / (2 * x)
a = (vf**2 - vi**2) / (2 * x)
print(f"The acceleration is {a:.2e}")
```

In Example 3.7, we investigated a jet landing on an aircraft
carrier. In a later maneuver, the jet comes in for a
landing on solid ground with a speed of 100 m/s, and
its acceleration can have a maximum magnitude of
5.00 m/s2 as it comes to rest. 

**(a)** 

From the instant the jet
touches the runway, what is the minimum time interval
needed before it can come to rest? 

```python
vi = 100 * (u.m / u.s)
vf = 0
a = -5.00 * (u.m / u.s**2)

# vf = vi + a * t ==> t = (vf - vi) / a
t = (vf - vi) / a
print(f"Minimun time: {t}")
```

**(b)** 

Can this jet land at a small tropical island airport where the runway is
0.800 km long? Explain your answer.

```python
x = vi * t + 1 / 2 * a * t**2
print(f"The minimun runaway is {x.to(u.km)}")
```

#### Problem 17

page 74

An object moving with uniform acceleration has a
velocity of 12.0 cm/s in the positive x direction when its
x coordinate is 3.00 cm. If its x coordinate 2.00 s later is
-5.00 cm, what is its acceleration?

```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
(x_i, x_f, v_i, t, a) = sp.symbols("x_i, x_f, v_i, t, a", positive=True)

# Define the kinematic equation: x_f = x_i + v_i * t + 1/2 * a * t^2
equation = sp.Eq(x_f, x_i + v_i * t + sp.Rational(1, 2) * a * t**2)

# Solve the equation for the unknown acceleration 'a'

solutions = sp.solve(equation, a)
a_isolated = solutions[0]

print("--- acceleration Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(a, a_isolated))}"))
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
v_i_val = 12.0 * (u.cm / u.s)
x_i_val = 3.00 * u.cm
t_val = 2.00 * u.s
x_f_val = -5.00 * u.cm

# Convert the SymPy expression to a callable function
a_func = sp.lambdify((x_i, x_f, v_i, t), a_isolated)

# Calculate numerical value
a_val = a_func(x_i_val, x_f_val, v_i_val, t_val)

print("--- AstroPy Numerical Result ---")
print(f"Acceleration a: {a_val}")
```
