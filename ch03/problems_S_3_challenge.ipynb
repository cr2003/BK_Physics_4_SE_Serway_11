Review. You are sitting in your car at rest at a traffic
light with a bicyclist at rest next to you in the adjoining
bicycle lane. As soon as the traffic light turns green,
your car speeds up from rest to 50.0 mi/h with constant
acceleration 9.00 mi/h/s and thereafter moves with
a constant speed of 50.0 mi/h. At the same time, the
cyclist speeds up from rest to 20.0 mi/h with constant
acceleration 13.0 mi/h/s and thereafter moves with a
constant speed of 20.0 mi/h. 

**(a)** For what time interval after the light turned green is 
the bicycle ahead of your car? 

**(b)** What is the maximum distance by which the
bicycle leads your car during this time interval?


```python
# ==============================================================================
# 1. SYMBOLIC STAGE (SymPy)
# ==============================================================================

# Define abstract mathematical symbols
(t, v_c_max, a_c, v_b_max, a_b) = sp.symbols("t v_c_max a_c v_b_max a_b", positive=True)

# Phase 1 duration for bicycle acceleration
t1_b = v_b_max / a_b

# Define position functions
x_bike = sp.Rational(1, 2) * a_b * t1_b**2 + v_b_max * (t - t1_b)
x_car = sp.Rational(1, 2) * a_c * t**2

# (a) Overtake Condition: x_car = x_bike
eq_overtake = sp.Eq(x_car, x_bike)
t_overtake_sol = sp.solve(eq_overtake, t)[1]  # Physical root

# (b) Max Separation Condition: v_car = v_bike_max
t_max_sep = v_b_max / a_c
max_sep_expr = sp.simplify(x_bike.subs(t, t_max_sep) - x_car.subs(t, t_max_sep))

print("--- Overtake Time Formula Solved by SymPy ---")
display(Math(rf"t = {sp.latex(t_overtake_sol)}"))

print("\n--- Maximum Lead Distance Formula Solved by SymPy ---")
display(Math(rf"\Delta x_{{\text{{max}}}} = {sp.latex(max_sep_expr)}"))
```

```python
# ==============================================================================
# 2. NUMERICAL EVALUATION WITH ASTROPY UNITS
# ==============================================================================

# Given parameters with AstroPy units
v_c_max_val = 50.0 * (u.imperial.mile / u.h)
a_c_val = 9.00 * (u.imperial.mile / u.h / u.s)

v_b_max_val = 20.0 * (u.imperial.mile / u.h)
a_b_val = 13.0 * (u.imperial.mile / u.h / u.s)

# Convert SymPy expressions to callable functions
t_func = sp.lambdify((v_c_max, a_c, v_b_max, a_b), t_overtake_sol)
max_sep_func = sp.lambdify((v_c_max, a_c, v_b_max, a_b), max_sep_expr)

# Calculate numerical values
t_val = t_func(v_c_max_val, a_c_val, v_b_max_val, a_b_val)
sep_val = max_sep_func(v_c_max_val, a_c_val, v_b_max_val, a_b_val)

print("--- AstroPy Numerical Results ---")
print(f"(a) Time interval bicycle is ahead: {t_val:.2f}")
print(
    f"(b) Maximum lead distance: {sep_val.to(u.imperial.ft):.1f} ({sep_val.to(u.m):.2f})"
)
```
