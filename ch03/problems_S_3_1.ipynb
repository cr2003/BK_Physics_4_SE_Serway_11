#### Problem 2

page 73

A particle moves according to the equation $x = 10t^2$, where
x is in meters and t is in seconds. 

**(a)** 

Find the average velocity for the time interval from 2.00 s to 3.00 s. 

```python
def f_x(t):
    x = 10 * t**2
    return x


ti = 2.00
tf = 3.00
xi = f_x(ti)
xf = f_x(tf)
v_avg = (xf - xi) / (tf - ti)
print(f"Average velocity for the interval {ti} - {tf} s = {v_avg} m/s")
```

**(b)** 

Find the average velocity for the time interval from 2.00 to 2.10 s.

```python
tf = 2.10
xf = f_x(tf)
v_avg = (xf - xi) / (tf - ti)
print(f"Average velocity for the interval {ti} - {tf} s = {v_avg:.1f} m/s")
```
