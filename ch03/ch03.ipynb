# 3 Motion in One Dimension
## 3.1 Velocity and Speed of a Particle

page 46

**Average Velocity**

$$v_{x,avg} \approxeq \frac{\Delta x}{\Delta t} \; \left[\frac{m}{s} \right]$$

**Average Speed**

$$v_{avg} \approxeq \frac{d}{\Delta t} \; \left[\frac{m}{s} \right]$$





## 3.2 Instantaneous Velocity and Speed

**Instantaneous Velocity**

$$v_{x} \approxeq \lim_{\Delta t \to 0} \frac{\Delta x}{\Delta t} = \frac{dx}{dt}$$

**Instantaneous Speed**

The instantaneous speed is defined as the magnitude of its istantaneous velocity.


## 3.3 Analysis Model: Particle Under Constant Velocity

$$v_x = \frac{\Delta x}{\Delta t}$$

$$\Delta x = x_f - x_i \implies v_x = \frac{(x_f - x_i)}{\Delta t}$$

$$ x_f = x_i + v_x \Delta t$$

In practice, we usually choose the time at the beginning of the interval to be $ t_i = 0$ and the
time at the end of the interval to be $ t_f = t$, so our equation becomes

$$ x_f = x_i + v_x t \;\;\;(\text{for constant}\; v_x)$$



## 3.5 Acceleration

page 57

**Average Acceleration**

$$a_{x,avg} \approxeq \frac{\Delta v_x}{\Delta t} = \frac{v_{xf} - v_{xi}}{t_{f} - t_{i}} ; \left[\frac{m}{s^2} \right]$$

**Instantaneous Acceleration**

$$a_{x} \approxeq \lim_{\Delta t \to 0} \frac{\Delta v_x}{\Delta t} = \frac{dv_x}{dt}$$

Because $v_x = \frac{dx}{dt}$

$$a_{x} = \frac{dv_x}{dt} = \frac{d}{dt} \left ( \frac{dx}{dt} \right) = \frac{d^2 x}{dt^2} $$
