Why is the following situation impossible? Emily challenges David to catch
a $1 bill as follows. She holds the bill vertically as shown in Figure P3.25,

#### Problem 26

An attacker at the base of a castle wall 3.65 m high
throws a rock straight up with speed 7.40 m/s from a
height of 1.55 m above the ground. 

**(a)** 
Will the rock
reach the top of the wall? 

**(b**) 

If so, what is its speed at
the top? If not, what initial speed must it have to reach
the top? 

```python
yf = 3.65 * u.m
vi = 7.40 * (u.m / u.s)
yi = 1.55 * u.m
g = -9.80 * (u.m / u.s**2)

# vf**2 = vi**2 + 2 * a * (yf - yi)
vf = (vi**2 + 2 * g * (yf - yi)) ** (1 / 2)
print(f"Yes, the rock will reach the top of the wall.")
print(f"With a speed of {vf:.2f}")
```

**(c)** Find the change in speed of a rock thrown
straight down from the top of the wall at an initial
speed of 7.40 m/s and moving between the same two
points. 

```python
yi = 3.65 * u.m
vi = -7.40 * (u.m / u.s)
yf = 1.55 * u.m
g = -9.80 * (u.m / u.s**2)


vf = (vi**2 + 2 * g * (yf - yi)) ** (1 / 2)
change_speed = vf - abs(vi)

print(f"The change in speed of the rock thrown down is {change_speed:.2f}")
```

#### Problem 29

page 76

A student throws a set of keys vertically upward to her
sorority sister, who is in a window 4.00 m above. The
second student catches the keys 1.50 s later. (a) With
what initial velocity were the keys thrown? (b) What
was the velocity of the keys just before they were caught?

```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define abstract mathematical symbols
(y, y_0, v_0, v, v_0, t, a) = sp.symbols("y, y_0, v_0, v, v_0, t, a", positive=True)

# Define the kinematic equation: y = y_0 + v_0 * t + 1/2 * a * t^2
equation = sp.Eq(y, y_0 + v_0 * t + sp.Rational(1, 2) * a * t**2)

# Solve the equation for the unknown velocity 'v_0'

solutions = sp.solve(equation, v_0)
v_0_isolated = solutions[0]

print("--- velocity Formula Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(v_0, v_0_isolated))}"))
```

```python
# -----------------------------------------------------------------
# 2. Numerical Evaluation with AstroPy Units
# -----------------------------------------------------------------

# Given parameters with AstroPy units
print("(a)")
t_val = 1.5 * u.s
y_val = 4 * u.m
y_0_val = 0 * u.m
g = -9.8 * (u.m / u.s**2)
a_val = g

# Convert the SymPy expression to a callable function
v_0_func = sp.lambdify((y, y_0, a, t), v_0_isolated)

# Calculate numerical value
v_0_val = v_0_func(y_val, y_0_val, a_val, t_val)

print("--- AstroPy Numerical Result ---")
print(f"Velocity v_0: {v_0_val:.1f}")
```

```python
print("(b)")
v_f_val = v_0_val + a_val * t_val
print(f"Final velocity v_f: {v_f_val:.2f}")
print(
    "The negative sign means that the keys are moving downward just before they are caught."
)
```
