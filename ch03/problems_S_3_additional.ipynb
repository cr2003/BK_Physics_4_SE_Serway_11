Review. As it passes over Grand Bahama Island, the
eye of a hurricane is moving in a direction 60.0 deg north
of west with a speed of 41.0 km/h. 

**(a)**

What is the unit-vector
expression for the velocity of the hurricane? It
maintains this velocity for 3.00 h, at which time the
course of the hurricane suddenly shifts due north,
and its speed slows to a constant 25.0 km/h. This
new velocity is maintained for 1.50 h.


```python
# --- Part (a): Velocity 1 ---
speed1 = 41.0
theta1 = vp.radians(180 - 60)  # 60 degrees North of West means 120 deg from x
v1 = vp.vec(speed1 * vp.cos(theta1), speed1 * vp.sin(theta1), 0)
v1
```

**(b)** 

What is the
unit-vector expression for the new velocity of the hurricane?

```python
# --- Part (b): Velocity 2 ---
speed2 = 25.0  # due North, only component y

v2 = vp.vec(0, speed2, 0)
v2
```

**(c)** 

What is the unit-vector expression for the
displacement of the hurricane during the first 3.00 h?

```python
# --- Part (c): Displacement 1 ---
delta_t1 = 3.00  # delta t1 3 horas
delta_r1 = v1 * delta_t1
delta_r1
```

**(d)** 

What is the unit-vector expression for the displacement
of the hurricane during the latter 1.50 h?

```python
# --- Part (d): Displacement 2 ---
delta_t2 = 1.50  # delta t2 1.50 horas
delta_r2 = v2 * delta_t2
delta_r2
```

**(e)** 

How far from Grand Bahama is the eye 4.50 h after it passes
over the island?

```python
# --- Part (e): Total Distance from the Origin ---
total_displacement = delta_r1 + delta_r2
distance = vp.mag(total_displacement)
distance
```

**Results**


```python
print("Results:")
print(f"(a) Velocity 1 (unit-vector): {fmt_vec(v1)} km/h")
print(f"(b) Velocity 2 (unit-vector): {fmt_vec(v2)} km/h")
print(f"(c) Displacement 1 (unit-vector): {fmt_vec(delta_r1)} km")
print(f"(d) Displacement 2 (unit-vector): {fmt_vec(delta_r2)} km")
print(f"(e) Final distance from Grand Bahama: {distance:.0f} km")
```

Inge run a race they invent. Inge runs directly from the
tail of the snake to its head, and Olaf starts from the same
place at the same moment but runs along the snake. 

**(a)** 

If both children run steadily at 12.0 km/h, Inge reaches the
head of the snake how much earlier than Olaf? 


```python
l = 420 * u.m  # length of the snake
theta = radians(180 + 105)  # from x-axis
s1 = 240 * u.m
s2 = l - s1
v = 12.0 * (u.km / u.h)
v = v.to(u.m / u.s)

# We choose the coordinates origin at the bent of the snake
# vector from origen to tail
vec_ot = vec(-s1.value, 0, 0)
# vector from origen to head
vec_oh = vec(s2.value * cos(theta), s2.value * sin(theta), 0)

# vector r from tail to head
vec_th = vec_oh - vec_ot

# length of the hypotenuse, straight from tail to head
hyp = vec_th.mag * u.m

# Inge time to run the hypotenuse
# v = s / t ==> t = s / v
t_inge = hyp / v
print(f"Inge runs along the hypotenuse = {hyp:.1f}")
print(f"Time of Inge = {t_inge:.1f}")

# Olaf time to run the whole length of the snake
t_olaf = l / v
print(f"Olaf runs the whole length of the snake = {l:.1f}")
print(f"Time of Olaf = {t_olaf:.1f}")

delta_t = t_olaf - t_inge
print(f"Inge reach the head earlier = {delta_t:.1f}")
```

**(b)** 

If Inge runs the race again at a constant speed of 12.0 km/h, at
what constant speed must Olaf run to reach the end of
the snake at the same time as Inge?

```python
# Olaf should run the whole length in the same time of Inge
v_olaf = l / t_inge
print(
    f"For Olaf to catch Inge, he should run at new velocity = {v_olaf:.2f} equivalent to {v_olaf.to(u.km / u.h):.1f}"
)
```

the box.

Why is the following situation impossible? A freight
train is lumbering along at a constant speed of 16.0 m/s.
Behind the freight train on the same track is a passenger
train traveling in the same direction at 40.0 m/s.
When the front of the passenger train is 58.5 m from
the back of the freight train, the engineer on the passenger
train recognizes the danger and hits the brakes
of his train, causing the train to move with acceleration
-3.00 m/s2. Because of the engineer’s action, the trains
do not collide.

```python
# To check for collision, we define the kinematics equations of move for both trains
# If there is an x equal for both at the same time, we have a collision

import sympy as sp

# 1. Define symbols
# t = time, x = position
# i1/i2 = initial position, v1/v2 = initial velocity, a1/a2 = acceleration
t = sp.symbols("t", real=True, positive=True)
xi1, vi1, a1 = sp.symbols("xi1, vi1, a1")
xi2, vi2, a2 = sp.symbols("xi2, vi2, a2")

# 2. Define the two equations
x1 = xi1 + vi1 * t + 0.5 * a1 * t**2
x2 = xi2 + vi2 * t + 0.5 * a2 * t**2

# 3. Solve for t where x1 == x2
equation = sp.Eq(x1, x2)
t_sol = sp.solve(equation, t)
t_sol
```

```python
# 4. Conver SymPy expression to a "Lambdified" function for numerical use
# This allows us to pass real numbers into the symbolic solution
collision_time_func = sp.lambdify((xi1, vi1, a1, xi2, vi2, a2), t_sol)

# 5. Plug in the train data
# Passenger: xi=0, vi=40, a=-3
# Freight: xi=58.5, vi=16, a=0
times = collision_time_func(0, 40, -3, 58.5, 16, 0)
for time in times:
    if time > 0:
        print(f"-> {time:.2f} s")
```

#### Problem 41

page 77

Hannah tests her new sports car by racing with Sam,
an experienced racer. Both start from rest, but Hannah
leaves the starting line 1.00 s after Sam does. Sam moves
with a constant acceleration of 3.50 m/s2, while Hannah
maintains an acceleration of 4.90 m/s2. Find 

**(a)**

the time at which Hannah overtakes Sam, 

```python
import sympy as sp

# 1. Define Symbols
# t is the time Sam has been traveling (Sam starts at t=0)
t = sp.symbols("t", real=True)

# Constants
a_sam = 3.50
a_hannah = 4.90
delay = 1.00

# 2. Define Position Equations
# x = x0 + v0*t + 0.5*a*t^2
# Sam starts at t=0
x_sam = 0.5 * a_sam * t**2
print(f"x_sam: {x_sam}")

# Hannah starts at t=1.0 so her 'active' time is (t - 1)
x_hannah = 0.5 * a_hannah * (t - delay) ** 2
print(f"x_hannah: {x_hannah}")


# 3. Solve for t (When x_sam == x_hannah)
equation = sp.Eq(x_sam, x_hannah)
print(f"equation: {equation}")
t_solutions = sp.solve(equation, t)
print(f"t_solutions: {t_solutions}")

# Filter for the solution where t > delay (Hannah must have started)
t_overtake = [sol for sol in t_solutions if sol > delay][0]
print("At the time of overtake:")
print(f"Time Sam has traveled: {t_overtake:.2f} s")
print(f"Time Hannah has traveled: {t_overtake - delay:.2f} s")
```

**(b)** 

the distance she travels before she catches him, and 

```python
# Calculate the Distance
distance = x_sam.subs(t, t_overtake)
print(f"Distance traveled: {distance:.0f} m")
```

**(c)** 

the speeds of both cars at the instant Hannah overtakes Sam.

```python
# Velocity equations: v = a * time
v_sam_final = a_sam * t_overtake
print(f"Sam's speed: {v_sam_final:.1f} m/s")

v_hannah_final = a_hannah * (t_overtake - delay)
print(f"Hannah's speed: {v_hannah_final:.1f} m/s")
```

Lisa rushes down onto a subway platform to find her
train already departing. She stops and watches the cars
go by. Each car is 8.60 m long. The first moves past
her in 1.50 s and the second in 1.10 s. Find the constant
acceleration of the train.

```python
import sympy as sp

# Define variables
vi, a, t1, t2, x1, x2 = sp.symbols("vi, a, t1, t2, x1, x2")

# x = vi*t + 1/2 * a * t**2
# Equation 1: First car
t1 = 1.50
t2 = 2.60  # (1.50 + 1.10)
x1 = 8.60
x2 = 8.60 * 2
eq1 = sp.Eq(vi * t1 + 0.5 * a * t1**2, x1)
eq2 = sp.Eq(vi * t2 + 0.5 * a * t2**2, x2)

# Solve the system
solution = sp.solve((eq1, eq2), (vi, a))

print(f"Initial Velocity (vi): {solution[vi]:.2f} m/s")
print(f"Constant Acceleration (a): {solution[a]:.2f} m/s^2")
```
