Suppose the position vector for a particle is given as a function
of time by $\vec{r}(t) = x(t)\hat{i} + y(t)\hat{j}$, with $x(t) = at + b$
and $y(t) = ct^2 + d$, where a = 1.00 m/s, b = 1.00 m,
c = 0.125 m/s2, and d = 1.00 m. 

**(a)**

Calculate the average velocity during the time interval from t = 2.00 s to
t = 4.00 s. 





```python
# 1. Symbolic Calculation with SymPy
t = sp.symbols("t")
a, b, c, d = 1.00, 1.00, 0.125, 1.0

# Define position components
x_expr = a * t + b
y_expr = c * t**2 + d

# Average Velocity from t=2 to t=4
# v_avg = (r(t2) - r(t1)) / t2 - t1
r_2 = np.array([float(x_expr.subs(t, 2)), float(y_expr.subs(t, 2)), 0])
r_4 = np.array([float(x_expr.subs(t, 4)), float(y_expr.subs(t, 4)), 0])
v_avg = (r_4 - r_2) / (4 - 2)
print(f"Average Velocity (2s to 4s): {v_avg} m/s")
```

**(b)** 

Determine the velocity and the speed at t = 2.00 s.


```python
# instantaneous velocity v(t) = dr/dt
v_x_expr = sp.diff(x_expr, t)
v_y_expr = sp.diff(y_expr, t)

vx_2 = float(v_x_expr.subs(t, 2))
vy_2 = float(v_y_expr.subs(t, 2))
v_instant = vec(vx_2, vy_2, 0)
speed_2 = sqrt(vx_2**2 + vy_2**2)

print(f"Velocity at 2s: {v_instant} m/s")
print(f"Speed at 2s: {speed_2:.3f} m/s")
```
