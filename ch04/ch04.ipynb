A long jumper (Fig. 4.12) leaves the ground at an angle of 20.0° above the horizontal
and at a speed of 11.0 m/s.

How far does he jump in the horizontal direction?

### Example 4.4
page 89

<img src="..//pics/Fig4_14.png" width=600> 

A stone is thrown from the top of a building upward at an angle of 30.0° to the horizontal with an initial speed of 20.0 m/s
as shown in Figure 4.14. The height from which the stone is thrown is 45.0 m above the ground.

**(A)** 

How long does it take the stone to reach the ground?


```python
import sympy as sp

# Define symbols
t, yf, yi, vi, g, theta = sp.symbols("t, y_f, y_i, v_i, g, theta")

# Define the position equation X = Xi + Vi * sin(theta) * T + 1/2 * A * t^2
equation = sp.Eq(yf, yi + vi * sp.sin(theta) * t + sp.Rational(1, 2) * g * t**2)
sp.pprint(equation)


# introduce the data
data = {
    yf: -45.0,  # m, origin at point of throwing
    yi: 0,
    vi: 20.0,  # m/s
    g: -9.8,  # m/s^2
    theta: radians(30),
}

# Substitute all at once
final_eq = equation.subs(data)
sp.pprint(final_eq)

solutions = sp.solve(final_eq, t)
for sol in solutions:
    if sol > 0:
        print(f"Time to reach the ground {sol:.2f} s")
```

**(B)** 

What is the speed of the stone just before it strikes the ground?

```python
theta = radians(30)
vi = 20.0 * (u.m / u.s)
vxi = vi * cos(theta)
vxf = vxi
vyi = vi * sin(theta)
g = -9.8 * (u.m / u.s**2)
t = 4.22 * u.s

vyf = vyi + g * t
# Final velocity is the sum of the x + y
vf = (vxf**2 + vyf**2) ** (1 / 2)
print(f"The final velocity of the stone is {vf:.1f}")
```

### Example 4.5
page 90

<img src="..//pics/Fig4_15.png" width=400> 

The End of the Ski Jump 

A ski jumper leaves the ski track moving in the horizontal direction with a speed of 25.0 m/s as shown in Figure 4.15. The landing incline below her falls off with a slope of 35.0°. Where does she land on the incline?


```python
import sympy as sp

# 1. Define symbols
d, phi, v_xi, g, t = sp.symbols("d, phi, v_xi, g, t")

# 2. Define the equations
# Origin beginning of the jump
# From the figure the right triangle shows that jumper's x and y coordinates at the landing point are given by
# xf = d cos(phi) and yf = -d sin(phi)
# xf = vxi*t ==> d cos(phi) = vxi * t  (2)
# yf = vyi * t - 1/2 * g * t^2 ==> -d sin(phi) = -1/2 * g * t^2 (4)

# Equation (2): x-coordinate of landing
eq_x = sp.Eq(d * sp.cos(phi), v_xi * t)
print("Equation 2: eq_x")
sp.pprint(eq_x)

# Equation (4): y-coordinate of landing, starting at y=0
eq_y = sp.Eq(-d * sp.sin(phi), -0.5 * g * t**2)
print("Equation 4: eq_y")
sp.pprint(eq_y)

# 3. Solve eq_x for t
t_expr = sp.solve(eq_x, t)[0]
print(f"Expression for t: {t_expr}")

# 4. Substitute t into eq_y
eq_y_substituted = eq_y.subs(t, t_expr)
print("eq_y_substituted by t:")
sp.pprint(eq_y_substituted)

# 5. solve for d
d_solution = sp.solve(eq_y_substituted, d)
# sp.solve returns a list; we take the non-zero solution
d_formula = d_solution[1]

print("\nDerived Formula for d:")
sp.pprint(d_formula)

# 6. Substitute Numerical Values
# v_xi = 25.0 m/s, phi = 35 degrees, g = 9.8 m/s^2
numerical_data = {
    v_xi: 25.0,
    phi: sp.rad(35.0),  # Convert degrees to radians
    g: 9.8,
}

d_value = d_formula.subs(numerical_data)
x_f = (d_value * sp.cos(phi)).subs(numerical_data)
y_f = (-d_value * sp.sin(phi)).subs(numerical_data)

# 7. Final Results
print(f"\n --- Numerical Results ---")
print(f"Distance (d): {d_value.evalf():.1f} m")
print(f"Landing x: {x_f.evalf():.1f} m")
print(f"Landing y: {y_f.evalf():.1f} m")
```

## 4.4 Analysis Model: Particle in Uniform Circular Motion
### Horizontal Range and Maximum Height of a Projectile

page 91

**Centripetal acceleration**

An acceleration of this nature is called a `centripetal acceleration` (centripetal means
center-seeking). 

$$a_c = \frac{v^2}{r}$$

**Period**

Is the time interval required for one complete revolution. That's it the number of seconds for one revolution.

$$T = \frac{2 \pi r}{v}$$

**Angular speed**
Because one full revolution of the particle around the circle corresponds to an angle of $2\pi$ radians, the product of $2\pi$ and the rotation rate gives the `angular speed` $\omega$ of the particle, measured in radians/s or $s^{-1}$

$$\omega = \frac{2\pi}{T}$$

Combining this equation with the equation of the period, we find a relationship between angular speed and the translational speed.

$$v = r\omega$$

We can express the centripetal acceleration of a particle in uniform circular motion in terms fo angular speed.

$$a_c = r \omega^2$$

### Example 4.6
The Centripetal Acceleration of the Earth

page 94

**(A)** 

What is the centripetal acceleration of the Earth as it moves in its orbit around the Sun?

```python
T = 1 * u.yr  # peoriod of Earth
T = T.to(u.s)
r = 1 * const.au  # radius of Earth's orbit around the Sun
r = r.to(u.m)

# a_c = v**2 / r
# v = (2*pi*r) / T
# ==> a_c = (4 * pi**2 * r) / T**2
a_c = (4 * pi**2 * r) / T**2
print(f"Earth's centripetal acceleration a_c={a_c:.2e}")
```

## 4.5 Tangential and Radial Acceleration

page 94

<img src="..//pics/Fig4_17.png" > 

As the particle moves along a curved path, the direction of the total acceleration vector $\vec{a}$ changes from point to point. At any instant, the vector can be resolved into two components based on an origin at the center of the dashed circle corresponding to that instant: a radial compoent $a_r$ along the radius of the circle and a tangential compoent $a_t$ perpendicular to the radius. The total acceleration vector $\vec{a}$  can be written as the vector sum of the component vectors:

$$\vec{a} = \vec{a}_r + \vec{a}_t$$

The tangential acceleration component causes a change in the speed $v$ of the particle. This component is parallel to the instantaneous velocity, and is given by

$$a_t = \frac{dv}{dt}$$

The radial acceleration component arises from a change in direction of the velocity vector and is given by

$$a_r = -a_c = -\frac{v^2}{r}$$

where $r$ is the radius of curvature of the path at the point in question.

Because $\vec{a}_r$ and $\vec{a}_t$ are perpendicular componentes of $\vec{a}$, it follows that the magnitude of $\vec{a}$ is

$$a = \sqrt{a_r^2 + a_t^2}$$


### Example 4.7
Over the Rise
page 94

<img src="..//pics/Fig4_18.png" width=400 > 

A car leaves a stop sign and exhibits a constant acceleration of 0.300 m/s2 parallel to the roadway. The car passes over a rise in the roadway such that the top of the rise is shaped like an arc of a circle of radius 500 m. At the moment the car is at the top of the rise, its velocity vector is horizontal and has a magnitude of 6.00 m/s. What are the magnitude and direction of the total acceleration vector for the car at this instant?


```python
# Tangential acceleration (magnitud) a_t = 0.300 m/s^2
a_t = 0.300 * (u.m / u.s**2)
print(f"Tangential acceleration a_t={a_t}")

# Radial acceleration (magnitud) a_r = - v^2 / r
r = 500 * u.m
v = 6.00 * (u.m / u.s)
a_r = -(v**2) / r
print(f"Radial acceleration a_r={a_r}")

# Total acceleration (magnitud) a = (a_r**2 + a_t**2)**(1/2)
a = (a_r**2 + a_t**2) ** (1 / 2)
print(f"Total acceleration a={a:.3f}")

# Angle phi=atan(a_r/a_t)
phi = degrees(atan(a_r / a_t))
print(f"The component of the angle respect to the horizontal phi={phi:.1f} deg")
```

**Finalize**

## 4.6 Relative Velocity and Relative Acceleration

page 96

<img src="..//pics/Fig4_21.png" width=300 > 


In this discussion of relative velocity, we use a double-subscript notation; the first subscript represents what is being observed, and the second represents who is doing the observing. Therefore, the notation $\vec{v}_{BA}$ means the velocity of observer B (and the attached frame $S_B$) as measured by observer A. With this notation, observer B measures A to be moving to the left with a velocity $\vec{v}_{AB} = -\vec{v}_{BA}$. For purpose of this discussion, let us place each observer at her or his repsective origin.

We define the time t = 0 as the instant at which the origins of the two reference frames coincide in space. Therefore, at later time t, the origins of the reference frames will be separated by a distance $\vec{v}_{BA} t$. We label the position P of the particle relative to observer A with the position vector $\vec{r}_{PA}$, and that relative to observer B with the position vector $\vec{r}_{PB}$, both at the time t. Figure 4.21 shows that the vectors $\vec{r}_{PA}$ and $\vec{r}_{PB}$ are related to each other through the expression

$$\vec{r}_{PA} = \vec{r}_{PB} + \vec{V}_{BA} t$$

$$\vec{u}_{PA} = \vec{u}_{PB} + \vec{V}_{BA} $$

These two equations are known as `Gallilean transformation equations`. They relate the position and velocity of a particle as measured by observers in relative motion.


### Example 4.8
A Boat Crossing a River

page 98

<img src="..//pics/Fig4_22.png" width=400 > 

A boat crossing a wide river moves with a speed of 10.0 km/h
relative to the water. The water in the river has a uniform
speed of 5.00 km/h due east relative to the Earth.
(A) If the boat heads due north, determine the velocity of
the boat relative to an observer standing on either bank.


```python
# Velocity of the boat respect to the river
u_br = 10.0 * (u.km / u.h)

# Velocity of the river water respect to the Earth (due east)
v_rE = 5.0 * (u.km / u.h)

u_bE = (u_br**2 + v_rE**2) ** (1 / 2)
print(f"Velocity relative to an observer at the river bank u_bE={u_bE:.1f}")

# Direction of the boat to the North (Angle theta of u_bE East of North)
# theta = arctan(v_rE / u_br)
theta = degrees(atan2(v_rE.value, u_br.value))
print(f"The boat direction east of north relative to Earth theta={theta:.1f} deg")
```

(B) If the boat travels with the same speed of 10.0 km(h relative to the river and is to travel due north as shown in Figure 4.22b, what should its heading be?

```python
# The new triangle show in Figure 4.22b, now the u_bE is not longer the hypotenuse, so the new equation is
u_bE = (u_br**2 - v_rE**2) ** (1 / 2)
print(f"Velocity relative to an observer at the river bank u_bE={u_bE:.2f}")

# Direction of the boat to the North (Angle phi of u_bE West of North)
# phi = arctan(v_rE / u_bE)
phi = degrees(atan2(v_rE.value, u_bE.value))
print(f"The boat direction west of north relative to Earth phi={phi:.1f} deg")
```

### Summary

#### displacement vector

$$\Delta \vec{r} = \vec{r_f} - \vec{r_i}$$

#### average velocity

$$\vec{v}_{avg} = \frac{\Delta \vec{r} }{\Delta t}$$


#### instantaneous velocity

$$\vec{v} \equiv \lim\limits_{\Delta t \to 0}\frac{\vec{r}}{\Delta t} = \frac{d\vec{r}}{dt}$$


#### average acceleration

$$\vec{a}_{avg} \equiv \frac{\Delta \vec{v} }{\Delta t} = \frac{\vec{v}_f - \vec{v}_i}{t_f - t_i}$$


#### instantaneous acceleration

$$\vec{a} \equiv \lim\limits_{\Delta t \to 0}\frac{\Delta\vec{v}}{\Delta t} = \frac{d\vec{v}}{dt}$$



#### Concepts and principles

$$\vec{v}_f = \vec{v}_i + \vec{a} t $$

$$\vec{r}_f = \vec{r}_i + \vec{v}_i t + \frac{1}{2} \vec{a} t^2 $$



#### frames of reference

$$\vec{u}_{PA} = \vec{u}_{PB} + \vec{v}_{BA}  $$


### Circular motion
#### Centripetal acceleration

$$a_c = \frac{\vec{v}^2}{r}$$

#### Period
$$T = \frac{2\pi r}{v}$$

#### angular speed
$$\omega = \frac{2\pi}{T}$$
