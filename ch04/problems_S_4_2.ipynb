```python
def fmt_vec(v, precision=2, type="f"):
    """Format the output of a vector with the specified precision"""
    return f"<{v.x:.{precision}{type}}, {v.y:.{precision}{type}}, {v.z:.{precision}{type}}>"
```

Review. A snowmobile is originally at the point with
position vector 29.0 m at 95.0° counterclockwise from the
x axis, moving with velocity 4.50 m/s at 40.0°. It moves
with constant acceleration 1.90 m/s2 at 200°. After 5.00 s
have elapsed, find (a) its velocity and (b) its position vector.



```python
r_i = vp.vec(29.0 * vp.cos(vp.radians(95)), 29.0 * vp.sin(vp.radians(95)), 0)
v_i = vp.vec(4.50 * vp.cos(vp.radians(40)), 4.50 * vp.sin(vp.radians(40)), 0)
a = vp.vec(1.90 * vp.cos(vp.radians(200)), 1.90 * vp.sin(vp.radians(200)), 0)

# a) v_f = v_i + a * t
t = 5
v_f = v_i + a * t
print(f"Velocity v_f = {fmt_vec(v_f)} m/s")
```

```python
# b) r_f = r_i + v_i * t + 1/2 + a * t**2
r_f = r_i + v_i * t + 1 / 2 * a * t**2

print(f"Position r_f = {fmt_vec(r_f)} m")
```
