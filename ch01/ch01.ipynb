## 1.4 Conversion of Units

```python
d_mile = 1 * u.imperial.mi
d = d_mile.to(u.m)
print(f"1 mile = {d} = {d.to(u.km):.3f}")

d2 = 1 * u.m
d2_in = d2.to(u.imperial.inch)
d2_ft = d2.to(u.imperial.ft)
print(f"1 m = {d2_in:.2f} = {d2_ft:.3f}")

d3 = 1 * u.imperial.ft
d3_m = d3.to(u.m)
print(f"1 ft = {d3_m} = {d3_m.to(u.cm)}")

d4 = 1 * u.imperial.inch
d4_m = d4.to(u.m)
print(f"1 in = {d4_m:.4f} = {d4_m.to(u.cm)}")
```

**Is He Speeding?**

On an interstate highway in a rural region of Wyoming, a car is traveling at a speed of 38.0 m/s. Is the driver exceeding the speed limit of 75.0 mi/h?

```python
v = 38 * (u.m / u.s)
v_mi_h = v.to(u.imperial.mi / u.h)
print(f"The car speed = {v_mi_h:.1f}")
print("The car is exceeding the spedd limit!!!")
```

**What If?**

What if the driver were from outside the United States and is familiar with

```python
print(f"The car speed = {v.to(u.km / u.h):.0f} ")
```
