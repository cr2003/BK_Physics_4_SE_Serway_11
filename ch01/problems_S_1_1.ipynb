Use information on the endpapers of this book to calculate the average density of the Earth.

**Answer:**

Modeling the Earth as a sphere, we find its volume as:

$$V = \frac{4}{3} \pi  r^3$$

The density $\rho$ (Greek letter rho) of any substance is defined as its mass per unit volume:

$$\rho \approxeq \frac{m}{V}$$

```python
m = c.M_earth
r = c.R_earth
V = 4 / 3 * pi * r**3
rho = m / V
print(f"Earth mass {m:.2e}")
print(f"Earth volume: {V:.2e}")
print(f"Earth density: {rho:.2e}")
```

Two spheres are cut from a certain uniform rock. One
has radius 4.50 cm. The mass of the other is five times
greater. Find its radius.

```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define symbols
# m1: mass of sphere 1, r1: radius of sphere 1
# m2: mass of sphere 2, r2: radius of sphere 2
m1, r1, r2 = sp.symbols("m1 r1 r2", positive=True)
m2 = 5 * m1  # Given: mass of the other is 5 times greater

# Set up the volumes
V1 = (4 / 3) * sp.pi * r1**3
V2 = (4 / 3) * sp.pi * r2**3

# Since density is uniform: m1 / V1 = m2 / V2
equation = sp.Eq(m1 / V1, m2 / V2)

# Solve the equation explicitly for r2
solutions = sp.solve(equation, r2)
r2_isolated = solutions[0]

print("--- Radius Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(r2, r2_isolated))}"))
```

```python
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (AstroPy)
# =====================================================================

# Assign the problem data using AstroPy units
r_small = 4.5 * u.cm  # radius of the small sphere

# Convert the SymPy expression into a callable Python function
# The order of the arguments matches the symbols we want to pass
calculate_radius = sp.lambdify((r1), r2_isolated)

# Evaluate the function passing the AstroPy Quantity objects directly
r_large = calculate_radius(r_small)

# Display the final result
print("--- Final Result ---")
print(f"Large sphere radius: {r_large:.2f}")
```

### Problem 4
What mass of a material with density $\rho$ is required to make a hollow spherical shell having inner radius r1 and outer radius r2?

#### Answer:
The volume of a spherical shell can be calculated from

$$V = V_0 - V_i = \frac{4}{3} \pi (r_2^3 - r_1^3)$$

From the definition of density, $\rho = \frac{m}{V}$, so

$$m = \rho V = \rho  \left( \frac{4}{3} \pi \right)  \left( r_2^3 - r_1^3 \right) = \frac{4 \pi \rho \left( r_2^3 - r_1^3 \right)}{3}$$

### Problem 5
You have been hired by the defense attorney as an expert witness in a lawsuit. The plaintiff is someone who just returned from being a passenger on the first orbital space tourist flight. Based on a travel brochure offered by the space travel company, the plaintiff expected to be able to see the Great Wall of China from his orbital height of 200 km above the Earth’s surface. He was unable to do so, and is now demanding that his fare be refunded and to receive additional financial compensation to cover his great disappointment. Construct the basis
for an argument for the defense that shows that his expectation of seeing the Great Wall from orbit was unreasonable. The Wall is 7 m wide at its widest point and the normal visual acuity of the human eye is $3\times 10^{-4}$ rad. (Visual acuity is the smallest subtended angle that an object can make at the eye and still be recognized; the subtended angle in radians is the ratio of the width of an object to the distance of the object from your eyes.)

```python
width = 7 * u.meter
distance = 200 * u.km
visual_acuity = 3e-4 * u.rad


# subtended angle
sub_angle = width / distance.to(u.meter) * u.rad

print(
    f"The subtended angle by the Great Wall is: {sub_angle}, that is {visual_acuity / sub_angle:.1f}"
)
print(f"times smaller than the visual acuity: {visual_acuity:.0e}, of the human eye.")
```
