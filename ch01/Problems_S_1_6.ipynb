### Problem 20
Book 10th Ed page 18, Instructor solution P1.36 page 15


#### Problem statement

How many significant figures are in the following numbers?

(a) $78.9 \pm  0.2 $

(b) $3.788 \times 10^9$

(c) $2.46 \times 10^{-6}$

(d) $0.0053$

#### Answer

(a) The $\pm 0.2$ express uncertainty in the last digit. Therefore, there are **three** significant digits in  $78.9 \pm 0.2$

(b) Scientific notation is often used to remove the ambiguity of the number of significant figures in a number. Therefore all the digits in $3.788$ are significant, and $3.788 \times 10^9$ has **four** significant figures.

(c) Similarly $2.46 \times 10^{-6}$ has **three** siginificant figures.

(d) Zeros used to position the decimal point are not significant. Therefore $0.0053$ has **two** siginificant figures.

The tropical year, the time interval from one vernal equinox to the next vernal equinox, is the basis for our calendar. It
contains 365.242 199 days. Find the number of seconds in a tropical year.

#### Answer

We work to nine significant digits

$$1 yr = 1 \; yr \left( \frac{365.242199 \; d}{1 \; yr}\right)

### 22. 
Review. The average density of the planet Uranus is
$1.27 \times 10^3 kg/m^3$. The ratio of the mass of Neptune to
that of Uranus is 1.19. The ratio of the radius of Neptune
to that of Uranus is 0.969. Find the average density of
Neptune.

$$\frac{M_N}{M_U} = 1.19 \; and \; \frac{r_N}{r_U} = 0.969$$
$$\rho = \frac{mass}{volume} = \frac{M}{V}$$
$$V = \frac{4}{3}\pi r^3$$
$$\frac{\rho_N}{\rho_U} = \frac{M_N/V_N}{M_U/V_U} = \left( \frac{M_N}{M_U}\right) \left( \frac{V_U}{V_N}\right)
= \left( \frac{M_N}{M_U}\right) \left( \frac{r_U}{r_N}\right)^3
$$

```python
rho_U = 1.27e3 * (u.kg / u.m**3)
ratio_M_N_by_N_U = 1.19
ratio_r_N_by_r_U = 0.969
ratio_rho_N_by_rho_U = ratio_M_N_by_N_U * (1 / ratio_r_N_by_r_U) ** 3
ratio_rho_N_by_rho_U

# ratio_rho_N_by_rho_U = rho_N / rho_U ==> rho_N = ratio_rho_N_by_rho_U * rho_U
rho_N = ratio_rho_N_by_rho_U * rho_U
print(f"Density of Neptune = {rho_N:.2e}")
```

### Problem 32
Book 10th Ed page 19, Instructor solution P1.58 page 23


#### Problem statement

Why is the following situation impossible? In an effort to boost interest in a television game show, each weekly winner is
offered an additional $1 million bonus prize if he or she can personally count out that exact amount from a supply of
one-dollar bills. The winner must do this task under supervision by television show executives and within one 40-hour
work week. To the dismay of the show’s producers, most contestants succeed at the challenge.

#### Answer

Assume the winner counts one dollar per second, and mantain the count without stopping. The time interval required for the task would be:

$$  \$10^6 \left( \frac{1\; s}{\$1}\right) 
            \left( \frac{1\; hour}{3600\;s}\right)
            \left( \frac{1\; work \; week}{40 \; hours}\right)
            = 6.9 \;work \; week$$



```python
# Define new units
dollar = u.def_unit("Dollar")
work_week = u.def_unit("WorkWeek", 40 * u.hour)

# Assume a person can count 1 dollar bill per second
count_rate = 1 * dollar / u.second


# Calculate the time it would take to count n dollars
def calculate_time(n):
    time = n / count_rate
    return time.to(work_week)


# Perform our problem
amount = 1e6 * dollar
print(f"It would take {calculate_time(amount):.2f} to count {amount:,.0f}.")
```

#### Problem statement

Bacteria and other prokaryotes are found deep underground, in water, and in the air. One micron ($10^{-6}$ m) is a
typical length scale associated with these microbes. 

(a) Estimate the total number of bacteria and other prokaryotes on the Earth. 

(b) Estimate the total mass of all such microbes. 

#### Answer assumptions:

typical length of bacterium: $L = 10^{-6} \; m$

typical volume of bacterium: $L^3 = 10^{-18} \; m^3$

surface area of Earth: $ A = 4 \pi r^2$

bacteria are found to a depth below Earth surface: $d = 1000 \; m$;  $V = A \times d$

average or 1000 bacteria in every $1 \; mm^3$

bacteria is basically composed of water, so we apply water density.

```python
# a
depth = 1000 * u.m
A = 4 * pi * c.R_earth**2
V = A * depth
bacteria_vol = 1e-18 * u.m**3
bacteria_density = 10**3 * u.kg / u.m**3
avg_bacteria_ratio = 1000 / (1 * u.mm**3)
num_bacteria = (avg_bacteria_ratio * V).decompose()

# b
mass_bacteria = num_bacteria * bacteria_density * bacteria_vol

print(f"The depth about: {depth:.0f}")
print(f"The Earth raius about: {c.R_earth:,.0f}")
print(f"The Area of Earth about: {A:.2e}")
print(f"The volume of Earth with bacteria about: {V:.2e}")
print(f"The bacteria volume about: {bacteria_vol}")
print(f"The bacteria density (same as water) about: {bacteria_density}")
print(f"The average bacteria ratio about: {avg_bacteria_ratio}")
print(f"The number of bacteria is about: {num_bacteria:.2e}")
print()
print(f"The mass of bacteria is about: {mass_bacteria:.2e}")
```

### Problem 39 
A woman wishing to know the height of a mountain
measures the angle of elevation of the mountaintop as
$12.0^\circ$. After walking 1.00 km closer to the mountain on
level ground, she finds the angle to be $14.0^\circ$. (a) Draw
a picture of the problem, neglecting the height of the
woman’s eyes above the ground. Hint: Use two triangles.
(b) Using the symbol y to represent the mountain
height and the symbol x to represent the woman’s
original distance from the mountain, label the picture.
(c) Using the labeled picture, write two trigonometric
equations relating the two selected variables. (d) Find
the height y.

#### Answer

$$\tan(12^\circ) = \frac{y}{x}$$

$$\tan(14^\circ) = \frac{y}{x - 1000}$$

1) Solve two equations for x

$$ x = \frac{y}{tan(12^\circ)}$$

$$ x = \frac{y}{tan(14^\circ)} + 1000$$

2) Setting the equations for x equal to each other and solving for y

$$y = \frac{1000}{1/tan(12^\circ) - 1/tan(14^\circ)}$$



```python
# Angles in degrees
a1 = radians(12.0)
a2 = radians(14.0)
dist = 1 * u.km
dist = dist.to(u.m)

# Formula derived: y = dist / (1/tan(a1) - 1/tan(a2))
y = dist / (1 / tan(a1) - 1 / tan(a2))
print(f"Mountain Height = {y:,.0f}")
```
