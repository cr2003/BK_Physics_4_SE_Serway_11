### Problem 11


A solid piece of lead has a mass of 23.94 g and a volume of 2.10 $cm^3$. From these data, calculate the density of lead in SI
units (kilograms per cubic meter).

$$\rho = \frac{m}{V}$$

```python
m_cm = 23.94 * u.g
m = m_cm.to(u.kg)
V_cm = 2.10 * u.cm**3
V = V_cm.to(u.m**3)
rho = m / V
print(f"Density of lead = {rho:.2e}")
```

### Problem 12 
Why is the following situation impossible? A student’s
dormitory room measures 3.8 m by 3.6 m, and its ceiling
is 2.5 m high. The student has a favorite author
and orders the author’s latest book from an online
retailer, with the intent to use the pages of the book
as wallpaper for the walls of their dorm room, including
the door and window. On the ordering website,
the book is listed with dimensions $8" \times 10"$ and is
546 pages long.

```python
L1 = 3.8 * u.m
L2 = 3.6 * u.m
H = 2.5 * u.m
l_in = 8 * u.imperial.inch
l = l_in.to(u.m)
h_in = 10 * u.imperial.inch
h = h_in.to(u.m)
book_pages = 546

walls_area = ((2 * L1) + (2 * L2)) * H
sheet_area = l * h
num_sheets = walls_area / sheet_area
num_pages = 2 * num_sheets
print(f"The required number of pages = {num_pages:,.0f}, book pages = {book_pages}")
```

### 13. 

One cubic meter (1.00 m3) of aluminum has a mass of
$2.70 \times 10^3 kg$, and the same volume of iron has a mass
of $7.86 \times 10^3 kg$. Find the radius of a solid aluminum
sphere that will balance a solid iron sphere of radius
2.00 cm on an equal-arm balance.

```python
# =====================================================================
# 1. SYMBOLIC STAGE (SymPy)
# =====================================================================

# Define positive symbols for densities and radii
rho_Al, rho_Fe = sp.symbols("rho_Al rho_Fe", positive=True)
r_Al, r_Fe = sp.symbols("r_Al r_Fe", positive=True)

# Set up masses based on density and spherical volume
m_Al = rho_Al * (4 / 3) * sp.pi * r_Al**3
m_Fe = rho_Fe * (4 / 3) * sp.pi * r_Fe**3

# Balance condition: mass of aluminum equals mass of iron
balance_eq = sp.Eq(m_Al, m_Fe)

# Solve explicitly for the radius of the aluminum sphere
solutions = sp.solve(balance_eq, r_Al)
r_Al_isolated = solutions[0]

print("--- Radius Solved by SymPy ---")
display(Math(rf"{sp.latex(sp.Eq(r_Al, r_Al_isolated))}"))
```

```python
# =====================================================================
# 2. UNITS AND NUMERICAL STAGE (AstroPy)
# =====================================================================

# Define given parameters with units
rho_Al_val = 2.70e3 * (u.kg / u.m**3)
rho_Fe_val = 7.86e3 * (u.kg / u.m**3)
r_Fe_val = 2.00 * u.cm

# Convert the SymPy expression into a callable Python function
# The order of the arguments matches the symbols we want to pass
calculate_radius = sp.lambdify((r_Fe, rho_Fe, rho_Al), r_Al_isolated)

# Evaluate the function passing the AstroPy Quantity objects directly
r_Al_val = calculate_radius(r_Fe_val, rho_Fe_val, rho_Al_val)

# Display the final result
print("--- Final Result ---")
print(f"Radius al sphere: {r_Al_val:.2f}")
```

### 15. 

One gallon of paint (volume $3.78 \times 10^{-3} m^3$) covers an
area of 25.0 m2. What is the thickness of the fresh paint
on the wall?

```python
V = 3.78e-3 * u.m**3
A = 25 * u.m**2

# V = A * thickness ==> thickness = V / A
thickness = V / A
print(f"The thickness = {thickness.to(u.mm):.3f} = {thickness.to(u.um):.0f}")
```

### 16. 

An auditorium measures $40.0 m \times 20.0 m \times 12.0 m$.
The density of air is 1.20 kg/m3. What are (a) the volume
of the room in cubic feet and (b) the weight of air in the
room in pounds?

```python
l = 40.0 * u.m
w = 20.0 * u.m
h = 12.0 * u.m
rho_air = 1.20 * (u.kg / u.m**3)

### (a)
V = l * w * h
print(f"The Volume = {V:,.0f} equivalent to {V.to(u.imperial.ft**3):,.0f}")

### (b)
# rho = mass / V ==> mass = rho * V
mass_air = rho_air * V
print(
    f"The mass of air = {mass_air:,.0f} equivalent to {mass_air.to(u.imperial.pound):,.0f}"
)
```
